## Coursework: Datascience Python libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing data and creating dataframes

```python
data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
data = pd.DataFrame(data)
data.set_index("Id", inplace = True)
test_data = pd.DataFrame(test_data)
test_data.set_index("Id", inplace = True)
data
```

## Data normalization

## NaN fields correction

```python
data.count()
```

```python
def null_fields_correction(fieldname):
    data.loc[data[fieldname].isnull(), fieldname] = data[fieldname].mean()
    test_data.loc[test_data[fieldname].isnull(), fieldname] = test_data[fieldname].mean()

null_fields_correction("LifeSquare")
null_fields_correction("Healthcare_1")
data
```

### Creating binary field from string columns Ecology_2, _ 3, Shops_2 

```python
data = pd.concat([data, pd.get_dummies(data['Ecology_2'], prefix="Ecology_2"), 
                  pd.get_dummies(data['Ecology_3'], prefix="Ecology_3"),
                  pd.get_dummies(data['Shops_2'], prefix="Shops_2")], axis=1)
data.drop(["Ecology_2", "Ecology_3", "Shops_2"], axis = 1, inplace = True)

test_data = pd.concat([test_data, pd.get_dummies(test_data['Ecology_2'], prefix="Ecology_2"), 
                  pd.get_dummies(test_data['Ecology_3'], prefix="Ecology_3"),
                  pd.get_dummies(test_data['Shops_2'], prefix="Shops_2")], axis=1)
test_data.drop(["Ecology_2", "Ecology_3", "Shops_2"], axis = 1, inplace = True)

data
```

## Preparing data for machine learning.

### Creating datasets for training and testing

```python
from sklearn.model_selection import train_test_split

y_train = data["Price"]
x_train = data.drop("Price", axis = 1)

x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, random_state = 42, test_size = 0.3)
```

### Researching the best parameters for Random Forest Regressor method

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import r2_score
from mpl_toolkits.mplot3d import Axes3D
arrx= []
arry = []
arrz = []
for i in range(5,20):
    for j in range(500,2001,500):
        model = RandomForestRegressor(n_estimators = j, max_depth = i, random_state = 42, n_jobs = -1)
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
        arrz.append(r2_score(y_test, y_pred))
        arrx.append(j)
        arry.append(i)
        
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(arrx, arry, arrz)
plt.show()
```

**As we see, number of n_estimators do not influence so much on the efficiency. Complitely different result for max_depth parametr which influences much more and the optimal one in my opinion is near 14**

```python
model = RandomForestRegressor(n_estimators = 1500, max_depth = 14, random_state = 42, n_jobs = -1)
model.fit(x_train, y_train)
```

**Getting final result**

```python
y_pred = model.predict(x_test)
r2_score(y_test, y_pred)
```

```python
y_pred = model.predict(test_data)
y_pred
```

## Exporting data into csv file

```python
y_pred = pd.DataFrame(y_pred, columns=["Price"])
test_data = test_data.reset_index()
y_pred = pd.concat([y_pred, test_data["Id"]], axis = 1)
y_pred = y_pred.set_index(["Id"])
y_pred
```
